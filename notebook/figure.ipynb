# 実行結果の可視化

## ライブラリのインポート

```python
import json
import matplotlib.pyplot as plt
```

## 結果を読み込み

```python
with open("../output/0.0_0.2.json", "r") as f:
    full_labeled = json.load(f)
with open("../output/0.5_0.1.json", "r") as f:
    unlabeled_05 = json.load(f)
with open("../output/0.9_0.01.json", "r") as f:
    unlabeled_09 = json.load(f)
```

## 表示

```python
full_labeled = [0] + full_labeled
unlabeled_05 = [0] + unlabeled_05
unlabeled_09 = [0] + unlabeled_09
```

```python
plt.ylim(-0.1, 1.1)
plt.plot(full_labeled[:8], label="full labeled")
plt.plot(unlabeled_05[:8], label="unlabeled 50%")
plt.plot(unlabeled_09[:8], label="unlabeled 90%")
plt.xlabel("epoch")
plt.ylabel("test accuracy")
plt.legend()
plt.show()
```
